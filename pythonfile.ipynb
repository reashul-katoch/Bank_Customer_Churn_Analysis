#creating the age group using the lables and qcuts

```python
lable=['young_adult','adult','mid_age','old_age']
df['age_group']=pd.qcut(df['Age'],q=4,labels=lable)
df.head(10)
```

```python
lablee=['new','repeat','old_customer']
df['tenure_group']=pd.qcut(df['tenure'],q=3,labels=lablee)
lab=['Poor','Fair','Good','Excellent']
df['credit_score_group']=pd.qcut(df['credit_score'],q=4,labels=lab)
df.head(10)
```

```python
df.rename(columns={'CustomerId':'customer_id','CreditScore':'credit_score','Geography':'geography','Gender':'gender','Age':'age','Tenure':'tenure',
'Balance':'balance','NumOfProducts':'num_of_products' ,'HasCrCard':'has_cr_card','IsActiveMember':'is_active_member','EstimatedSalary':'estimated_salary'  ,
               'Exited':'exited'},inplace=True)
```

```python
mapping={
    1:"yes",
    0:'no'
}
df['active_Status']=df['is_active_member'].map(mapping)
df['credit_cr_status']=df['has_cr_card'].map(mapping)
df.drop(['has_cr_card','is_active_member'],axis=1,inplace=True)
df.head()
```
